```python
print("привет")
print("чтобы выйти, напиши 'exit'\n")

start = input("Введите начало региона: ")
if start == 'exit':
    print("пока")
else:
    end = input("Введите конец региона: ")
    if end == 'exit':
        print("пока")
    else:
        start = int(start)
        end = int(end)
        
        if start > end:
            start, end = end, start
            
        print(f"\nЗагадай число от {start} до {end}, а я угадаю")
        
        while True:
            mid = (start + end) // 2
            answer = input(f"Твое число <, > или = {mid}? ")
            
            if answer == 'exit':
                print("пока")
                break
            elif answer == '=':
                print(f"Это было число {mid}!")
                break
            elif answer == '<':
                end = mid - 1
            elif answer == '>':
                start = mid + 1
                
            if start > end:
                print("Чисел в этом диапазоне больше нет")
                break
```
