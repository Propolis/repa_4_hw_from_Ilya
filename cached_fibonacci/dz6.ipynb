```python
import time

def fib_regular(n):
    if n <= 1:
        return n
    return fib_regular(n - 1) + fib_regular(n - 2)

def fib_cached(n, cache={}):
    if n in cache:
        return cache[n]
    if n <= 1:
        return n
    
    cache[n] = fib_cached(n - 1, cache) + fib_cached(n - 2, cache)
    return cache[n]

for n in [10, 20, 30, 60]:
    print(f"\n--- Тест для N = {n} ---")
    
    # Вызов с кэшем
    start_time = time.time()
    res_cache = fib_cached(n)
    time_cache = time.time() - start_time
    print(f"С кэшем:     F({n}) = {res_cache}, Время: {time_cache:.6f}сек")
    
    # Вызов без кэша только если N <= 30
    if n <= 30:
        start_time = time.time()
        res_reg = fib_regular(n)
        time_reg = time.time() - start_time
        print(f"Без кэша:    F({n}) = {res_reg}, Время: {time_reg:.4f}сек")
    else:
        print(f"Без кэша:    Пропускаем (N > 30, будет вычисляться слишком долго)")
```
