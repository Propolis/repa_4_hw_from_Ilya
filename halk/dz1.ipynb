```python
def request_validator(question, expected_answers=False):
    user_response = input(question).lower()
    print(question, '\n', user_response)
    if expected_answers:
        if not user_response in expected_answers:
            return request_validator(
                question=f'Ответа {user_response} нет, но есть такие:\n{expected_answers}\nОтветь на вопрос используя ответы выше.\n{question}',
                expected_answers=expected_answers)
        return user_response

    if not user_response.isdigit():
        return request_validator(
            question=f'Ответа {user_response} нет, можно ввести только цифру\nОтветь на вопрос еще раз.\n{question}')
    return int(user_response)
```

```python
if request_validator("хочешь сняться в кино?", ('да', 'нет')) == 'да':
    if request_validator("Халком или Локи - кем будешь ?", ('халком', 'халк', 'локи')) in ('халком', 'халк'):
        print('принят') if request_validator('сколько бицуха в объеме?') >= 60 else print('потеряйся')
    else:
        print('иди домой мальчик') if request_validator('маму или папу больше любишь?', ('маму', 'папу')) == 'папу' else print('Спроси у папы кого он больше любит')
else:
    print('потеряйся')
```
