```python
def print_pascal_triangle(n):
    row = [1]
    for i in range(n + 1):        
        spaces = "  " * (n - i)
        content = "   ".join(map(str, row))
        print(spaces + content)
        
        next_row = [1]
        for j in range(len(row) - 1):
            next_row.append(row[j] + row[j+1])
        next_row.append(1)
        row = next_row

print_pascal_triangle(4)
```
